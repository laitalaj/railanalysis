# Extracting data

You'll want to run this cell once before running any further cells.
It will extract the compressed data for you.

```python
from os import path
import py7zr

data_dir = 'data'
source_path = path.join(data_dir, 'compressed', 'nerails.7z')
network_path = path.join(data_dir, 'northeuroperails.graphml')

if not path.exists(network_path):
    print('Decompressing data...')
    compressed_data = py7zr.SevenZipFile(source_path)
    compressed_data.extractall(path = data_dir)
    print('Done!')
else:
    print(f'Data already decompressed at {network_path}')
```

# Loading data with [osmnx](https://github.com/gboeing/osmnx)

## Configure osmnx

```python
import osmnx as ox

ox.config(log_console = True, data_folder = 'data')
```

## Load data

This will take a while, as our data is quite big

```python
graph = ox.load_graphml('northeuroperails.graphml')
```

## Display data

```python
ox.plot_graph(graph)
```
